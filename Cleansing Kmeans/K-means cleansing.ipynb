```python
na_values = ['NO CLUE', 'N/A', '0','']
df = pd.read_csv('complete.csv', error_bad_lines=False,na_values=na_values)
```

```python
numdf = (df.drop('latitude', axis=1)
         .join(df['latitude'].apply(pd.to_numeric, errors='coerce')))

fix = (numdf.drop('longitude', axis=1)
         .join(numdf['longitude'].apply(pd.to_numeric, errors='coerce')))
```

```python
new = fix[['latitude', 'longitude']].copy()
```

```python
newnonan = new.dropna(how='any')
```

```python
from numpy import array
from math import sqrt
from pyspark.mllib.clustering import KMeans, KMeansModel
```

```python
from pyspark.sql import SQLContext
sqlCtx = SQLContext(sc)
frame = sqlCtx.createDataFrame(newnonan)
```

```python
frame.take(50)
```

```python
rdd = frame.rdd
```

```python
datasaved = rdd.map(lambda line: array([x for x in line]))
```

```python
datasaved.collect()
```

```python
clusters = KMeans.train(datasaved, 2, maxIterations=10000, initializationMode="random")
```

```python
# Evaluate clustering by computing Within Set Sum of Squared Errors
def error(point):
    center = clusters.centers[clusters.predict(point)]
    return sqrt(sum([x**2 for x in (point - center)]))

WSSSE = datasaved.map(lambda point: error(point)).reduce(lambda x, y: x + y)
print("Within Set Sum of Squared Error = " + str(WSSSE))
```
